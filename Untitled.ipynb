```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Covid-19(clss).csv")
# print(df)
df.head(10)
```

```python
df['Gender']=pd.get_dummies(df['Gender'])
df.head(10)
```

```python
# print(df.info())
print(df['Gender'].value_counts())
print(df['Fever'].value_counts())
print(df['Cough'].value_counts())
print(df['Runny_nose'].value_counts())
print(df['Age'].value_counts())
print(df['Breathing_problem'].value_counts())
print(df['Headache'].value_counts())
print(df['Pneumonia'].value_counts())
print(df['Travelling_in_last_15days'].value_counts())
print(df['Victims'].value_counts())
```

```python
# df.describe()
y=df[['Victims']]
X=df.drop('Victims',axis=1)
X.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(x_train.head())
print(y_train.head())
```
